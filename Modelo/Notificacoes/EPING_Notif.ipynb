```python
df = df.iloc[:,[0,1,4,6,7,10]]
```

```python
df.rename({df.columns[0]:"Country", df.columns[1]:"Country Code", df.columns[2]:"Year", df.columns[3]:"Commodity Code"},axis=1, inplace=True)
```

```python
df["Year"] = df["Year"].str.extract("\d{2}/\d{2}/(\d{2})")[0]
```

```python
df["Year"] = pd.to_numeric(df["Year"])
df["Year"] = df["Year"].apply(lambda x: x + 1900 if x > 20  else x + 2000)
```

```python
# df["Year"].unique()
df
```

```python
# df.set_index("Year", inplace=True)
df.sort_index(inplace=True)
```

```python
df = df.dropna()
```
