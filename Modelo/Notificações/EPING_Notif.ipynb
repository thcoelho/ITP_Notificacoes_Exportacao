```python
import pandas as pd
```

```python
df = pd.read_csv("EPingNotifications.csv", error_bad_lines=False)
df
```

```python
df
```

```python
df.columns
```

```python
df = df.iloc[:,[0,1,4,6,7,10]]
```

```python
df.rename({df.columns[0]:"Country", df.columns[1]:"Country Code", df.columns[2]:"Year", df.columns[3]:"Commodity Code"},axis=1, inplace=True)
```

```python
df = df.dropna()
```

```python
df["Year"] = df["Year"].apply(lambda x: str(x)[-2:])
df["Year"] = pd.to_numeric(df["Year"])
```

```python
df["Year"] = df["Year"].apply(lambda x: x + 1900 if x > 20  else x + 2000)
```

```python
df
```

```python
df.to_csv("EPING_Notifications.csv")
```
