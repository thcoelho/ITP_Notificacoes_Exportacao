```python
# Importar bibliotecas
import pandas as pd
```

```python
# Transformar arquivos CSV em DataFrames e concatená-los
def concatenar():
    
    df9698 = pd.read_csv("Exportações/Exportacoes_desagregadas_1996-1998.csv")
    df9903 = pd.read_csv("Exportações/Exportacoes_desagregadas_1999-2003.csv")
    df0408 = pd.read_csv("Exportações/Exportacoes_desagregadas_2004-2008.csv")
    df0913 = pd.read_csv("Exportações/Exportacoes_desagregadas_2009-2013.csv")
    df1418 = pd.read_csv("Exportações/Exportacoes_desagregadas_2014-2018.csv")
    
    lista = [df9698, df9903, df0408, df0913, df1418]
    
    df = pd.concat(lista)
    df = df.set_index("Year")
    df = df.sort_index()
    
    return df

df = concatenar()
df
```

```python
def filtro(df):
    lista = ['Partner', 'Partner ISO', 
         'Commodity Code', 'Commodity',
         'Trade Value (US$)']
    df = df[lista]
    df = df[df["Commodity Code"] != 16]
    return df

df = filtro(df)
df
```

```python
df.to_csv("Valor_Exportacao_Especificado.csv")
```
